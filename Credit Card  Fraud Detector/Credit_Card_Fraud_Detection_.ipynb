# Under-Sampling

Build a simple dataset containing similar distrubtion of normal and fraudulent trasaction

Number of Fraudlent Transactions - 492

```python
Legit_Sample=Legit.sample(n=492)   # Ot will take the Ramdom data set
```

# Concateninting Two data frames

```python
new_df=pd.concat([Legit_Sample,Fraud],axis=0) # Rowwise addition
```

# Splilting the data into Features & Targets

```python
X=new_df.drop(columns='Class',axis =1)
Y=new_df['Class']
```

```python
X.head(5)
```

```python
Y.head(5)
```

# Split the data into Training Data & Testing Data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.20,random_state=2)
```

```python
print(X.shape,xtrain.shape,xtest.shape)
```

```python
# Model Training  Logistics Regression
model =LogisticRegression()
```

```python
# Training the Logistics Regression Model with Trainig Data
```

```python
model.fit(xtrain,ytrain)
```

# Evalution

```python
# Accuracy Score
# Accuracy on Training data
xtrain_pred=model.predict(xtrain)
training_data_accuracy= accuracy_score(xtrain_pred,ytrain)
```

```python
print('Accuracy on Training Data : ',training_data_accuracy)
```

```python
# Accuracy on Test data
xtest_pred=model.predict(xtest)
test_data_accuracy= accuracy_score(xtest_pred,ytest)
```

```python
print('Accuracy on Test Data : ',test_data_accuracy)
```
